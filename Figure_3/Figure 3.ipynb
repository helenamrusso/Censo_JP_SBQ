```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
import matplotlib.patches as Wedge
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
```

```python
# Load and process data
df = pd.read_csv('Results_clear.txt', sep='\t')
df = pd.DataFrame(df['1.14_work_sector'].value_counts()).reset_index()
df.columns = ['1.14_work_sector', 'count']  # Renaming columns for clarity

# Define a custom autopct function to show percentages in parentheses
def autopct_format(pct):
    return f'({pct:.1f}%)'

# Create a donut chart
labels = df['1.14_work_sector']
sizes = df['count']

# Use Seaborn to get the 'tab20' color palette
colors = sns.color_palette("tab20", len(df))

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, wedgeprops=dict(width=0.3), startangle=90, autopct='%1.1f%%', colors=colors)

# Draw a circle in the center to make it a donut
center_circle = plt.Circle((0, 0), 0.70, color='white')
fig.gca().add_artist(center_circle)

# Equal aspect ratio ensures the pie is drawn as a circle.
ax.axis('equal')

# Display the chart
# plt.savefig('/.../Figure_3.pdf')
plt.show()
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
import matplotlib.patches as Wedge
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
```

# Trabalhou n√£o pago

```python
# Load and process data
df = pd.read_csv('Results_clear.txt', sep='\t')
df = pd.DataFrame(df['1.8_worked_unpaid'].value_counts()).reset_index()
df.columns = ['1.8_worked_unpaid', 'count']  # Renaming columns for clarity

# Add a new column for the percentage
df['percentage'] = ((df['count'] / df['count'].sum()) * 100).round(1)

# Custom function to display percentage and count
def autopct_format(pct):
    total = df['count'].sum()
    count = int(pct * total / 100.0)
    return f'{pct:.1f}% ({count+1})'

# Use Seaborn to get the 'tab20' color palette
colors = sns.color_palette("Pastel1", len(df))

# Create pie chart with black edges and custom labels
plt.figure(figsize=(8, 6))
plt.pie(
    df['percentage'], 
    labels=df['1.8_worked_unpaid'], 
    autopct=autopct_format, 
    startangle=140,
    colors=colors,  # Apply tab20 color palette
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}  # Set edge color and width
)


# plt.savefig('/.../Figure_2a.pdf')
plt.show()
```

```python
df = pd.read_csv('inflation_increase_scholarship.txt', sep='\t')

# Initialize the plot
plt.figure(figsize=(12, 6))

# Plotting Inflation (cumulative) with dots for each year
ax = sns.lineplot(data=df, x='year', y='Inflation (cumulative)', marker='o', label='Inflation (Cumulative)',
                 color = 'cornflowerblue')
plt.fill_between(df['year'], df['Inflation (cumulative)'], alpha=0.05, color='cornflowerblue')  # Pale blue fill

# Plotting Scholarship (increase) with dots for each year
sns.lineplot(data=df, x='year', y='Scholarship (increase)', marker='o', label='Scholarship (Increase)',
            color = 'purple')
plt.fill_between(df['year'], df['Scholarship (increase)'], alpha=0.1, color='purple')  # Pale blue fill


# Adding titles and labels
plt.xlabel('Ano')
plt.ylabel('Porcentagem de aumento')
plt.legend()

# Customize plot appearance
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Ensure all years are shown on the x-axis
plt.xticks(df['year'], rotation=45)  # Rotate for readability if needed

# plt.savefig('/.../Figure_2c.pdf')
plt.show()
```
